## Asignación 2 &#x1F625;

Utilizando las características __Pclass__, __Sex__, __SibSp__, __Parch__, __Fare__, __Cabin__ y __Embarked__, entrene un clasificador lineal para predecir __Survived__ utilizando el algoritmo _pocket_. Imprima el error obtenido.

```python
x = df.drop(['PassengerId', 'Survived', 'Name', 'Age', 'Ticket'], axis='columns').values
y = df.Survived.values
```

```python
def train_pla(
    data,
    target,
    max_iters=1000
) -> np.array: 
    """
    Descripción:
    Esta funcion ejecuta una variante del algoritmo PLA para clasificar datos linealmente separables.
    
    Entradas:
    - data: Indica las características utilizadas para la clasificación.
    - target: Indíca las etiquetas que se esperan obtener despues de una correcta clasificación.
    - max_iters: Indica el número máximo de iteraciones del algoritmo. Si el algoritmo no converge
                antes del número máximo de iteraciones, se detiene y devuelve los parámetros 
                calculados hasta el momento.
            
    Salidas:
    - Un arreglo de numpy indicando los parámetros calculados
    """
        
    # model = __random_fn__((data.shape[1] + 1,))
    
    # CÓDIGO AQUÍ
    model = np.random.rand(data.shape[1]+ 1)
    
    X = np.insert(data, 0, 1, axis=1)
    y = target.copy().reshape(-1, 1)

    

    iters = 0
    while iters < max_iters:
        prediction = np.sign(X @ model)
        misclassified = (prediction != y).squeeze()
        misclassified_indices = np.where(misclassified)[0]

        # misclassified = misclassified.reshape(-1)
        if not misclassified.any():
            break

        misclassified_X = X[misclassified_indices]
        misclassified_y = y[misclassified_indices]
        
        model += (misclassified_y * misclassified_X).sum(axis=0)
        iters += 1
        
    return model
```

```python
def _error_rate(X, y, w):
    X_new = np.hstack((np.ones((X.shape[0], 1)), X))
    y_pred = np.sign(np.dot(X_new, w))
    return np.mean(y_pred != y)
```

```python
def pocket_algorithm(X, y, max_iterations = 3):
    # Agregamos una columna de unos a las datos de entrada para tener en cuenta el sesgo
    X = np.insert(X, 0, 1, axis=1)
    
    # Inicializamos el vector de pesos
    w = np.zeros(X.shape[1] + 1)
    
    # Inicializar el mejor vector de peso y tasa de error
    best_w = []#w#np.copy(w)
    best_error_rate = 10000000
    
    # Inicializamos el pocket algorithm 
    for t in range(max_iterations):
        print(f'Iteracion: {t}')
        # ejecutar PLA hasta que no existan más puntos mal clasificados
        w = train_pla(X, y)
        
        # evaluate Ein(w(t + 1))
        aux = _error_rate(X, y, w)
        # print(aux, best_error_rate)
        best_w.append([aux, w])
        # check if w(t + 1) is better than the current best in terms of Ein
        if aux < best_error_rate:
            # best_w.= aux
            best_error_rate = aux
            
    print(best_w[1:], best_w[0])
    return [best_w[1:], best_w[0]] # return the weights without the bias term and the bias term
```

## Asignación 3 &#x1F625;

Utilizando las características __Pclass__, __Sex__, __SibSp__, __Parch__, __Cabin__ y __Embarked__, entrene una regresión lineal para predecir __Fare__ utilizando el algoritmo de Ordinary Leasts Squares (OLS). Imprima el valor del error cuadrático medio (MSE).

```python
def mean_squared_error(y_true, y_pred):
    """
    Calcula el error cuadrático medio entre los objetivos verdadero y previsto
    
    Entradas
    y_true: Vector objetivo verdadero
    y_pred: Vector objetivo predicho
    
    Salida:
    Error cuadrático medio
    """
    return round(np.mean((y_true - y_pred) ** 2), 1)

def OLS(X, y):
    """
    Calcula los pesos óptimos utilizando mínimos cuadrados ordinarios
    :param X: Matriz de características de forma (n_muestras, n_características)
    :param y: Vector objetivo de la forma (n_muestras,)
    :return: Pesos óptimos de la forma (n_características,)
    """
    
    # Calculamos la pseudo-inversa de x
    x_t = ((np.linalg.pinv(x) @ x) ** -1) @ np.linalg.pinv(x)
    
    # Calculamos el vector de pesos
    w = x_t @ y
    
    # Realizamos las predicciones del modelo
    y_pred = np.dot(x, w)
    
    print(f'Error cuadratico medio: {mean_squared_error(y, y_pred)}')
    return w
```
